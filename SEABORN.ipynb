```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

var=[1,2,3,4,5,5,9]
var1=[2,3,4,5,6,3,7]
x1=pd.DataFrame({"var":var,"var1":var1})
sns.lineplot(x="var",y="var1",data=x1)
plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```
